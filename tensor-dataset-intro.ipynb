```python
#train test split using PYTORCH and DATALOADER
from torch.utils.data import TensorDataset,DataLoader
```

```python
df
```

```python
data = df.drop('target',axis=1).values
labels = df['target'].values
```

```python
iris = TensorDataset(torch.Tensor(data),torch.LongTensor(labels))
```

```python
type(iris) #its a TensorDataset object
```

```python
len(iris)
```

```python
for i in iris:
    print(i)
```

```python
iris_loader = DataLoader(iris,batch_size=10,shuffle = True)
#divide the data in batches and shuffles it internally
```

```python
for i_batch,sample_batch in enumerate(iris_loader):
    #iris tensordataset object as per batches(#serial number of the the batch,batch data)
    print(i_batch,sample_batch)
```

```python
#generally keeping track of batch number is not important so we use:
for batch in iris_loader:
    print(batch)
```
